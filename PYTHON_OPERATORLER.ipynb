```python
#klavyeden sıfır girilinceye kadar 0 girilmesini sağlayan ve bu sayıların kaç tane olduğu, sayıların toplamı ve ortalama sonuçlarını ekrana yazan program

sayi=1
s=0
toplam=0
while sayi!=0:
  sayi=int(input("bir sayi giriniz:"))
  if sayi!=0:
    toplam+=sayi
    s+=1
  if sayi==0:
    print("sayi girişi yapmadınız")
else:
  ortalama=toplam/s
  print(s,"tane sayi girilmiştir")
  print(toplam)
  print(ortalama)
```
